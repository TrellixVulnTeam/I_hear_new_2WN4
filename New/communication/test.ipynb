```python
import time
frame_list = []
frame_list.append([[], [], []])
startTime = time.time()*1000
camera = cv2.VideoCapture(0)
while((time.time()*1000 - startTime) < 8150):
    ret, frame = camera.read()
    if(ret == True):
        frame_list[len(frame_list) - 1][2].append(frame)
```
